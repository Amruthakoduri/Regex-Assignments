## Assignment-6 EDA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('bmh')
```

```python
data  = pd.read_csv('creditcard.csv') 
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
data.head(10).T
```

```python
data.describe()
```

```python
data.info()
```

```python
print(data['Amount'].describe())
plt.figure(figsize=(7, 5))
plt.scatter(data['Amount'],data['Class'],color='Blue')
plt.show()
```

```python
print(data['Time'].describe())
plt.figure(figsize=(7, 5))
plt.scatter(data['Time'],data['Amount'],color='Blue')
plt.show()
```

```python
plt.figure(dpi=100)
sns.countplot(x='Class', data = data)
```

```python
plt.figure(dpi=100)
sns.boxplot(y="Amount", data = data)
```

```python
plt.figure(dpi=100)
sns.boxplot(x="Time", data = data)
```

```python
plt.figure(dpi=120)
plt.grid(color="black")
sns.violinplot(y="Time", data = data)
```

### Numerical data distribution

```python
list(set(data.dtypes.tolist()))
```

```python
data_num = data.select_dtypes(include = ['float64', 'int64'])
data_num.head()
```

```python
data_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8); # ;
```

```python
df_num_corr = data.corr()['Amount'][:-2]
features_list = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with Amount:\n{}".format(len(features_list), features_list))
```

```python
for i in range(0, len(data.columns), 5):
    sns.pairplot(data=data,
                x_vars=data.columns[i:i+5],
                y_vars=['Amount'])
```

```python
corr = data.drop('Amount', axis=1).corr() # We already examined SalePrice correlations
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```
